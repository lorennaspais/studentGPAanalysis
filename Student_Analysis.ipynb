##INFO DATASET

Student ID

StudentID: A unique identifier assigned to each student (1001 to 3392).

Demographic Details


*   Age: The age of the students ranges from 15 to 18 years.
*   Gender: Gender of the students, where 0 represents Male and 1 represents Female.
*   Ethnicity: The ethnicity of the students, coded as follows:
        0: Caucasian
        1: African American
        2: Asian
        3: Other
*   ParentalEducation: The education level of the parents, coded as follows:
        0: None
        1: High School
        2: Some College
        3: Bachelor's
        4: Higher

Study Habits

*   StudyTimeWeekly: Weekly study time in hours, ranging from 0 to 20.
*   Absences: Number of absences during the school year, ranging from 0 to 30.
*   Tutoring: Tutoring status, where 0 indicates No and 1 indicates Yes.

Parental Involvement

*   ParentalSupport: The level of parental support, coded as follows:
        0: None
        1: Low
        2: Moderate
        3: High
        4: Very High

Extracurricular Activities

*   Extracurricular: Participation in extracurricular activities, where 0 indicates No and 1 indicates Yes.
*   Sports: Participation in sports, where 0 indicates No and 1 indicates Yes.
*   Music: Participation in music activities, where 0 indicates No and 1 indicates Yes.
*   Volunteering: Participation in volunteering, where 0 indicates No and 1 indicates Yes.

Academic Performance

*   GPA: Grade Point Average on a scale from 2.0 to 4.0, influenced by study habits, parental involvement, and extracurricular activities.
*   Target Variable: Grade Class
*   GradeClass: Classification of students' grades based on GPA:
        0: 'A' (GPA >= 3.5)
        1: 'B' (3.0 <= GPA < 3.5)
        2: 'C' (2.5 <= GPA < 3.0)
        3: 'D' (2.0 <= GPA < 2.5)
        4: 'F' (GPA < 2.0)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Student_performance_data _.csv')
df.describe()
```

```python
df.head()
```

```python
#Substituindo Etnias

# Define a mapping for the ethnicity codes
ethnicity_mapping = {
    0: 'Caucasian',
    1: 'African American',
    2: 'Asian',
    3: 'Other'
}

# Apply the mapping to the 'Ethnicity' column
df['Ethnicity'] = df['Ethnicity'].map(ethnicity_mapping)

# Display the first few rows to confirm the changes
df.head()
```

```python
#Substituindo parental education

# Define a mapping for the parental education codes
parentaleducation_mapping = {
    0: 'High School',
    1: 'Some College',
    2: 'Bachelor',
    3: 'Master',
    4: 'Higher'
}

# Apply the mapping to the 'Parental Education' column
df['ParentalEducation'] = df['ParentalEducation'].map(parentaleducation_mapping)


# Display the first few rows to confirm the changes
df.head()
```

```python
#Substituindo parental education

# Define a mapping for the parental education codes
gender_mapping = {
    0: 'Male',
    1: 'Female'
}

# Apply the mapping to the 'Parental Education' column
df['Gender'] = df['Gender'].map(gender_mapping)


# Display the first few rows to confirm the changes
df.head()
```

```python
# Converter os valores de GPA da escala de 0 a 4 para a escala de 0 a 100
df['GPA (0-100)'] = df['GPA'] * 25

# Mostrar as primeiras linhas para verificar a converso
df
```

```python

# Agrupar os dados em faixas de tempo de estudo semanal
df['StudyTimeWeeklyBins'] = pd.cut(df['StudyTimeWeekly'], bins=5)

# Ordenar os intervalos e convert-los para strings
df['StudyTimeWeeklyBins'] = df['StudyTimeWeeklyBins'].apply(lambda x: f"{x.left}-{x.right}" if pd.notnull(x) else "NaN")

# Calcular a mdia do GPA para cada faixa de tempo de estudo semanal
gpa_by_study_time = df.groupby('StudyTimeWeeklyBins')['GPA (0-100)'].mean().reset_index()

# Renomear as colunas para melhor entendimento
gpa_by_study_time.columns = ['Faixa de Tempo de Estudo Semanal', 'GPA Mdio']

# Ordenar o DataFrame resultante pelos intervalos
gpa_by_study_time = gpa_by_study_time.sort_values(by='Faixa de Tempo de Estudo Semanal')

# Criar e exibir o grfico de linhas
grafico_linhas_gpa_tempo = px.line(gpa_by_study_time, x='Faixa de Tempo de Estudo Semanal', y='GPA Mdio', title='GPA por Tempo de Estudo')
grafico_linhas_gpa_tempo.show()
```

```python
# Calcular a correlao entre GPA e tempo de estudo semanal
correlation = df['GPA (0-100)'].corr(df['StudyTimeWeekly'])

# Exibir a correlao
print("Correlao entre GPA e Tempo de Estudo Semanal:", correlation)

# Interpretao da correlao


if correlation >= 0.7:
    interpretation = "Alta correlao positiva"
elif 0.5 <= correlation < 0.7:
    interpretation = "Correlao moderada positiva"
elif 0.3 <= correlation < 0.5:
    interpretation = "Correlao baixa positiva"
elif -0.3 < correlation < 0.3:
    interpretation = "Muito baixa ou sem correlao"
elif -0.5 < correlation <= -0.3:
    interpretation = "Correlao baixa negativa"
elif -0.7 < correlation <= -0.5:
    interpretation = "Correlao moderada negativa"
else:
    interpretation = "Alta correlao negativa"

print("Interpretao:", interpretation)
```
